```python
import numpy as np
import pandas as pd

%precision 3
```

```python
df = pd.read_csv('../stat_analysis/data/ch2_scores_em.csv', index_col='student number')
```

```python
en_scores = np.array(df['english'])[:10]
ma_scores = np.array(df['mathematics'])[:10]

scores_df = pd.DataFrame({'english':en_scores,
                         'mathematics':ma_scores},
                         index = pd.Index(['A','B','C','D','E','F','G','H','I','J'],
                         name ='student'))
scores_df

```

```python
summary_df = scores_df.copy()
summary_df['english_deviation'] =\
    summary_df['english'] - summary_df['english'].mean()
summary_df['mathematics_deviation']=\
    summary_df['mathematics'] - summary_df['mathematics'].mean()
summary_df['product of deviation'] =\
    summary_df['english_deviation'] * summary_df['mathematics_deviation']
summary_df
```

```python
import matplotlib.pyplot as plt

%matplotlib inLine
```

```python
english_scores = np.array(df['english'])
math_scores = np.array(df['mathematics'])

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111)
#산점도
ax.scatter(english_scores, math_scores)
ax.set_xlabel('english')
ax.set_ylabel('mathematics')

plt.show()
```

1.2.2. 회귀직선

```python
poly_fit = np.polyfit(english_scores, math_scores, 1) # bo, b1을 구함
poly_1d = np.poly1d(poly_fit) #bo+b1*x를 반환하는 함수를 작성

xs = np.linspace(english_scores.min(), english_scores.max()) #직선을 그리기 위해 x좌표 생성
ys = poly_1d(xs) # xs에 대응하는 y좌표 생성

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111)
#산점도
ax.scatter(english_scores, math_scores)
ax.set_xlabel('english')
ax.set_ylabel('mathematics')
ax.plot(xs, ys, color = 'gray',
       label = f'{poly_fit[1]:.2f} + {poly_fit[0]:.2f}x')
ax.legend(loc = 'upper left') # 범례 표시

plt.show()
```

```python
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111)

c = ax.hist2d(english_scores, math_scores,
              bins = [9,8], range = [(35, 80), (55, 95)])

ax.set_xlabel('english')
ax.set_ylabel('mathematics')
ax.set_xticks(c[1])
ax.set_yticks(c[2])
fig.colorbar(c[3], ax = ax) # 컬러 바 표시

plt.show()
```

```python
# npy 형식으로 저장된 Numpy array를 읽어드린다
anscombe_data = np.load('../stat_analysis/data/ch3_anscombe.npy')
print(anscombe_data.shape)
anscombe_data[0]
```

```python
stats_df = pd.DataFrame(index = ['X_mean', 'X_variance', 'Y_mean',
                                  'Y_variance', 'X&Y_correlaton', 'X&&Y_regression line'])
for i, data in enumerate(anscombe_data):
    dataX = data[:, 0]
    dataY = data[:, 1]
    poly_fit = np.polyfit(dataX, dataY, 1)
    stats_df[f'data{i+1}'] =\
        [f'{np.mean(dataX):.2f}', f'{np.var(dataX):.2f}',
         f'{np.mean(dataY):.2f}', f'{np.var(dataY):.2f}',
         f'{np.corrcoef(dataX, dataY)[0,1]:.2f}',
         f'{poly_fit[1]:.2f} + {poly_fit[0]:.2f}x']

stats_df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 그래프를 그리기 위한 영역을 2x2개 생성
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10),
                         sharex=True, sharey=True)

xs = np.linspace(0, 30, 100)
for i, data in enumerate(anscombe_data):
    # 1차 다항식(직선)으로 회귀선 구하기
    poly_fit = np.polyfit(data[:, 0], data[:, 1], 1)
    poly_1d = np.poly1d(poly_fit)
    ys = poly_1d(xs)

    # 그리는 영역 선택
    ax = axes[i // 2, i % 2]
    ax.set_xlim([4, 20])
    ax.set_ylim([3, 13])

    # 타이틀 추가
    ax.set_title(f'data{i+1}')

    # 산점도 및 회귀선 그리기
    ax.scatter(data[:, 0], data[:, 1])
    ax.plot(xs, ys, color='gray')

# 그래프 사이의 간격을 좁힘
plt.tight_layout()
plt.show()
```
